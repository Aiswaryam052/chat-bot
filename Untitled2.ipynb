```python
import nltk
from nltk.chat.util import Chat, reflections
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
import random
import datetime

# Download required NLTK data
nltk.download('punkt')
nltk.download('vader_lexicon')

pairs = [
    [
        r".*(hi|hello|hey|greetings).*",
        ["Hello! Welcome to College Admission Helpdesk. How can I assist you today?", 
         "Hi there! Ready to help with your college queries. What do you need?", 
         "Greetings! What admission information are you looking for?"]
    ],
    [
        r".*(what|which)?\s*(courses|programs|degrees)\s*(are)?\s*(available|offered|offer)?.*",
        ["We offer undergraduate and graduate programs in:\n"
         "- Computer Science\n- Engineering\n- Business\n- Arts & Humanities\n"
         "- Medicine\n- Law\n",
         ]
    ],
    [
        r".*(how|can I|to)?\s*(apply|application process|applying).*",
        ["The application process involves:\n"
         "1. Create an online account\n2. Complete application form\n"
         "3. Submit required documents\n4. Pay application fee\n"
         "5. Track your application status\n\nWould you like detailed steps?",
         "You can apply through our online portal or in-person."]
    ],
    [
        r".*(admission requirements|eligibility|requirements).*",
        ["General requirements include:\n"
         "- Completed application form\n- Official transcripts\n"
         "- Standardized test scores\n- Letters of recommendation\n"
         "- Personal statement\n\nSpecific programs may have additional requirements."]
    ],
    [
        r".*(tuition|fee|cost|price|how much).*",
        ["Tuition varies by program:\n- Undergraduate: $15,000–$25,000 per year\n"
         "- Graduate: $20,000–$35,000 per year\n"
         "Financial aid is available. Would you like information about scholarships?"]
    ],
    [
        r".*(scholarship|financial aid|funding|grant|aid).*",
        ["We offer several financial aid options:\n"
         "- Merit-based scholarships\n- Need-based grants\n"
         "- Work-study programs\n- External funding options\n\n"
         "Would you like to check eligibility criteria?"]
    ],
    
    [
        r".*(campus tour|visit).*",
        ["You can schedule a campus tour through our website. "
         "We offer:\n- Virtual tours\n- In-person visits\n- Open houses\n\n"
         ]
    ],
    [
        r".*(thank you|thanks|appreciate).*",
        ["You're welcome! Feel free to ask if you have more questions.",
         "Happy to help! Good luck with your application!",
         "My pleasure. Let me know if you need anything else!"]
    ],
    [
        r".*(quit|bye|exit).*",
        ["Thank you for chatting with us! Wishing you the best with your college journey.",
         "Goodbye! Hope to see you on campus soon!",
         "All the best with your application process!"]
    ]
]

# Sentiment analysis function
def analyze_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(text)
    if sentiment['compound'] >= 0.05:
        return "positive"
    elif sentiment['compound'] <= -0.05:
        return "negative"
    else:
        return "neutral"

# Correct spelling mistakes in user input using TextBlob
def correct_spelling(text):
    blob = TextBlob(text)
    corrected = str(blob.correct())
    return corrected

# Chatbot class
class CollegeAdmissionChatbot:
    def __init__(self):
        self.chatbot = Chat(pairs, reflections)
        
    def respond(self, user_input):
        # Correct spelling first
        corrected_input = correct_spelling(user_input.lower().strip())
        
        # Sentiment analysis
        sentiment = analyze_sentiment(corrected_input)
        
        # Get chatbot response based on corrected input
        response = self.chatbot.respond(corrected_input)
        
        if not response:
            return self.handle_unknown_query()
        
        # Sentiment-aware follow-up
        if sentiment == "positive":
            response += random.choice([
                "\n\nIs there anything more I can help with?", 
                "\n\nGlad to help! What else would you like to know?"])
        elif sentiment == "negative":
            response += "\n\nI'm sorry if this wasn't helpful. Could you rephrase your question?"
        
        return response

    def handle_unknown_query(self):
        return ("I'm not sure I understand your question. I'm trained to help with college admission queries. "
                "You can ask about:\n"
                "- Application process\n- Available programs\n- Tuition fees\n"
                "- Admission requirements\n- Campus visits")

def chat_with_bot():
    print("=" * 50)
    print("🎓 COLLEGE ADMISSION CHATBOT".center(50))
    print("=" * 50)
    print("\nHi! I'm your College Admission Assistant. Type 'quit' to exit.\n")
    print("📌 You can ask me things like:")
    print("  - What programs do you offer?")
    print("  - How do I apply?")
    print("  - What are the admission requirements?")
    print("  - What is the tuition fee?")
    print("  - Tell me about scholarships.")
    print("  - Can I visit the campus?")
    print("  - Thank you / Bye\n")

    chatbot = CollegeAdmissionChatbot()

    while True:
        try:
            user_input = input("You: ")
            if user_input.lower() in ['quit', 'exit', 'bye']:
                print("\nBot:", chatbot.respond(user_input))
                break
            response = chatbot.respond(user_input)
            print("Bot:", response)
        except KeyboardInterrupt:
            print("\n\nSession ended. Goodbye!")
            break
        except Exception as e:
            print("\nBot: Sorry, something went wrong. Could you please rephrase?")

if __name__ == "__main__":
    chat_with_bot()
```
